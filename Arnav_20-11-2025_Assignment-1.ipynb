### Question 61  Remove the nth word from a string.  

```python
s = input("Enter a string: ")
n = int(input("Enter the position of the word to remove: "))

words = s.split()          

if 1 <= n <= len(words):   
    words.pop(n - 1)      
    result = " ".join(words)
    print("Result:", result)
else:
    print("Invalid")
```

### Question 62 Insert a word at the nth position in a string. 

```python
s = input("Enter the original string: ")
word = input("Enter the word to insert: ")
n = int(input("Enter the position to insert at: "))

words = s.split()     

if 1 <= n <= len(words) + 1:
    words.insert(n - 1, word)    
    result = " ".join(words)
    print("Result:", result)
else:
    print("Invalid position!")
```
